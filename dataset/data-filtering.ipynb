```python
import json
```

```python
filename = 'drug-label-0001-of-0010.json'
file = open(filename)
data = json.load(file)
data = data['results']
print(len(data))
```

```python
drugs = []
acne, cough, diarrhea = 0, 0, 0

for i in range(len(data)): 
    if len(data[i]['openfda']) != 0: 
        
        if data[i]['openfda']['product_type'] == ['HUMAN OTC DRUG']:
            try: 
                purpose = data[i]['purpose'][0].lower()
                
                if 'acne' in purpose or 'cough' in purpose or 'diarrhea' in purpose:
                    try:                        
                        indications_and_usage = data[i]['indications_and_usage'][0].lower()
                        
                        if 'acne' or 'cough' or 'diarrhea' in indications_and_usage: 
                            print('brand_name:', data[i]['openfda']['brand_name'][0])
                            print('purpose:', purpose)
                            print('indication_and_usage:', indications_and_usage)
                            print()
                            
                            if 'acne' in indications_and_usage: 
                                data[i]['category'] = 'acne'
                                acne += 1
                            elif 'cough' in indications_and_usage:
                                data[i]['category'] = 'cough'
                                cough += 1
                            elif 'diarrhea' in indications_and_usage:
                                data[i]['category'] = 'diarrhea'
                                diarrhea +=1 
                                
                            drugs.append(i)
                            
                    except: 
                        pass
            except: 
                pass
```

```python
len(drugs), acne, cough, diarrhea
```

```python
fix_drugs = {}

for i in range(len(drugs)):
    fix_drugs[i] = data[drugs[i]]
```

```python
len(fix_drugs)
```

```python
with open("bangkit_0323_dataset.json", "w") as outfile: 
    json.dump(fix_drugs, outfile, indent=4)
```
